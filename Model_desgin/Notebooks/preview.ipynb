```python
from datasets import load_dataset, load_from_disk
import pandas as pd
from pathlib import Path
from rich.console import Console
from rich.table import Table
from IPython.display import display, Markdown
import random
console = Console()
```

```python
data_folder = Path("../Datasets/non_aug")   # ← change if needed

dataset = load_dataset(
    "csv",
    data_files={
        "train": str(data_folder / "train.csv"),
        "validation": str(data_folder / "val.csv"),
        "test": str(data_folder / "test.csv"),
    }
)
```

```python
console.print("\n[bold cyan]Dataset loaded:[/bold cyan]")
console.print(dataset)
```

```python

stats_table = Table(title="Dataset Statistics", show_header=True, header_style="bold magenta")

stats_table.add_column("Split", style="cyan", no_wrap=True)
stats_table.add_column("Rows", justify="right")
stats_table.add_column("Avg text length", justify="right")
stats_table.add_column("0 / 1 balance", justify="right")

for split in dataset:
    ds = dataset[split]
    count = len(ds)
    if count == 0:
        continue
    lengths = [len(ex["text"]) for ex in ds]
    avg_len = sum(lengths) / count
    label_counts = pd.Series(ds["label"]).value_counts(normalize=True).round(3)
    balance_str = f"{label_counts.get(0, 0.0):.3f} / {label_counts.get(1, 0.0):.3f}"
    
    stats_table.add_row(
        split,
        f"{count:,}",
        f"{avg_len:.0f}",
        balance_str
    )

console.print(stats_table)
```

```python

SAMPLES_PER_SPLIT = 5

def show_random_samples(split_name, n=5):
    ds = dataset[split_name]
    if len(ds) == 0:
        console.print(f"[yellow]No data in {split_name}[/yellow]")
        return
    
    indices = random.sample(range(len(ds)), min(n, len(ds)))
    
    console.rule(f"Random samples from {split_name.upper()} (showing {len(indices)})", style="green")
    
    for idx in indices:
        ex = ds[idx]
        label_str = "REAL" if ex["label"] == 0 else "FAKE"
        color = "green" if ex["label"] == 0 else "red"
        
        console.print(f"\n[bold]{label_str}[/bold]  (row {idx:,})")
        console.print(f"Text length: {len(ex['text']):,} chars")
        console.print(ex["text"][:600] + ("..." if len(ex["text"]) > 600 else ""))
        console.print("-" * 100)


for split in ["train", "validation", "test"]:
    if split in dataset:
        show_random_samples(split, SAMPLES_PER_SPLIT)
```

```python
def show_extremes(split_name, n=3):
    ds = dataset[split_name]
    if len(ds) == 0:
        return
    
    lengths = [(i, len(ex["text"])) for i, ex in enumerate(ds)]
    lengths.sort(key=lambda x: x[1])
    
    console.rule(f"Shortest {n} in {split_name.upper()}", style="blue")
    for idx, length in lengths[:n]:
        ex = ds[idx]
        label_str = "REAL" if ex["label"] == 0 else "FAKE"
        console.print(f"[{label_str}] len={length:4d} → {ex['text'][:180]}...")
    
    console.rule(f"Longest {n} in {split_name.upper()}", style="blue")
    for idx, length in lengths[-n:]:
        ex = ds[idx]
        label_str = "REAL" if ex["label"] == 0 else "FAKE"
        console.print(f"[{label_str}] len={length:6,d} → {ex['text'][:180]}...")


for split in ["train", "validation", "test"]:
    if split in dataset:
        show_extremes(split)
```

```python
# ────────────────────────────────────────────────────────────────
# Cell: Aggressive final cleaning + overwrite original CSV files
# ────────────────────────────────────────────────────────────────

import re
from rich.console import Console

console = Console()

# ────────────────────────────────────────────────────────────────
# Aggressive junk patterns (expand as needed)
# ────────────────────────────────────────────────────────────────

JUNK_PATTERNS = [
    r'(?i)before it\'?s news',
    r'(?i)beforeitsnews',
    r'(?i)ye olde false flag',
    r'(?i)%.*think.*fact.*add your.*two cents',
    r'(?i)add your two cents',
    r'(?i)read more at.*(false|flag|hoax)',
    r'(?i)source:.*http',
    r'(?i)cnnn',                    # common typo/spam
    r'(?i)\[.*\]|\{.*\}',           # bracket spam
    r'(?i)http.*\s*$',              # trailing links
    r'(?i)click here|subscribe|share this',
]

def aggressive_clean(text):
    if not isinstance(text, str):
        return " "
    
    # Basic normalization (from earlier)
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)
    text = re.sub(r'\n{2,}', '\n', text)
    text = re.sub(r'\t+', ' ', text)
    text = re.sub(r'\s{2,}', ' ', text)
    
    # Remove aggregator signatures & poll junk
    for pat in JUNK_PATTERNS:
        text = re.sub(pat, ' ', text)
    
    # Collapse repeated identical lines (simple heuristic)
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    unique_lines = []
    seen = set()
    for line in lines:
        if line not in seen:
            seen.add(line)
            unique_lines.append(line)
    
    text = '\n'.join(unique_lines)
    
    # Final cleanup
    text = re.sub(r'\s{2,}', ' ', text).strip()
    
    return text if text else " "


# ────────────────────────────────────────────────────────────────
# Apply aggressive cleaning to all splits
# ────────────────────────────────────────────────────────────────

console.rule("Aggressive cleaning — final pass", style="bold red")

cleaned_rows = 0

for split in dataset:
    if split not in ["train", "validation", "test"]:
        continue
    
    console.print(f"\n[bold cyan]Processing {split.upper()} ({len(dataset[split]):,} rows)[/bold cyan]")
    
    def clean_row(example):
        example["text"] = aggressive_clean(example["text"])
        return example
    
    # Update in place
    dataset[split] = dataset[split].map(
        clean_row,
        desc=f"Cleaning {split}",
        load_from_cache_file=False
    )
    
    # Optional: drop rows that became empty after aggressive clean
    before = len(dataset[split])
    dataset[split] = dataset[split].filter(
        lambda x: len(x["text"].strip()) >= 30,
        desc=f"Remove empty after aggressive clean - {split}"
    )
    removed = before - len(dataset[split])
    if removed > 0:
        console.print(f"  Removed {removed:,} empty/noisy rows")
    
    cleaned_rows += len(dataset[split])

console.print(f"\n[bold green]Total rows after aggressive clean:[/bold green] {cleaned_rows:,}")


# ────────────────────────────────────────────────────────────────
# Overwrite original CSV files in the same folder
# ────────────────────────────────────────────────────────────────

console.rule("Saving updated files (overwriting originals)", style="bold yellow")

folder = Path("../Datasets/non_aug")   # ← change if your folder has different name

for split in ["train", "validation", "test"]:
    if split not in dataset or len(dataset[split]) == 0:
        console.print(f"[yellow]Skipping {split} — no data[/yellow]")
        continue
    
    path = folder / f"{split}.csv" if split != "validation" else folder / "val.csv"
    
    df = dataset[split].to_pandas()[["text", "label"]]
    df.to_csv(path, index=False, encoding="utf-8")
    
    console.print(f"[green]Updated:[/green] {path}  →  {len(df):,} rows")

console.print("\n[bold magenta]Finished.[/bold magenta] Original files have been overwritten with aggressively cleaned versions.")
```

```python
# ────────────────────────────────────────────────────────────────
# Cell: Re-preview random samples AFTER aggressive cleaning
# ────────────────────────────────────────────────────────────────

console.rule("RE-PREVIEW AFTER AGGRESSIVE CLEANING", style="bold green")

SAMPLES_PER_SPLIT = 6   # you can increase to 8 or 10 if you want to see more

for split in ["train", "validation", "test"]:
    if split not in dataset or len(dataset[split]) == 0:
        console.print(f"[yellow]No data in {split}[/yellow]")
        continue
    
    console.print(f"\n[bold cyan]Random samples from {split.upper()} ({len(dataset[split]):,} rows total)[/bold cyan]")
    
    # Pick random indices
    indices = random.sample(range(len(dataset[split])), min(SAMPLES_PER_SPLIT, len(dataset[split])))
    
    for idx in sorted(indices):
        ex = dataset[split][idx]
        label_str = "REAL" if ex["label"] == 0 else "FAKE"
        color = "green" if ex["label"] == 0 else "red"
        
        preview = ex["text"][:700] + ("..." if len(ex["text"]) > 700 else "")
        length = len(ex["text"])
        
        console.print(f"\n[{color}]{label_str}[/{color}]  (row {idx:,} • {length:,} chars)")
        console.print(preview)
        console.print("─" * 100)
```
