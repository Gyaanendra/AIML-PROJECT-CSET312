```python
import pandas as pd
import numpy as np
from tqdm import tqdm
csv_path = r'../Datasets/meraged/final_combined_corpus.csv'
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
df.info()
```

```python
df['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Map labels to 0/1 if needed (real=0, fake=1 or vice versa)
df['label'] = df['label'].map({'real': 0, 'fake': 1})

# Stratified split: first train + temp, then split temp into val/test
train_df, temp_df = train_test_split(
    df, test_size=0.2, stratify=df['label'], random_state=42
)

val_df, test_df = train_test_split(
    temp_df, test_size=0.5, stratify=temp_df['label'], random_state=42
)

print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
print("Label ratios preserved:", train_df['label'].value_counts(normalize=True))
```

```python
train_df.to_csv('../Datasets/non_aug/train.csv', index=False)
val_df.to_csv('../Datasets/non_aug/val.csv', index=False)
test_df.to_csv('../Datasets/non_aug/test.csv', index=False)
```
